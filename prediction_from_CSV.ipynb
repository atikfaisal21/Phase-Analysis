## Choose classification algorithms

### Subtask:
Select a diverse set of binary classification algorithms, including traditional models (like Logistic Regression, Support Vector Machines, Random Forest, Gradient Boosting) and deep learning models (like a simple Feedforward Neural Network).


**Reasoning**:
Select and list the binary classification algorithms to be used for evaluation, including traditional and deep learning models.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasClassifier

# List of chosen algorithms
chosen_algorithms = {
    "Logistic Regression": LogisticRegression(random_state=42),
    "Support Vector Machine": SVC(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Feedforward Neural Network": KerasClassifier(
        model=Sequential([
            Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
            Dense(32, activation='relu'),
            Dense(1, activation='sigmoid')
        ]),
        loss='binary_crossentropy',
        optimizer='adam',
        metrics=['accuracy'],
        epochs=10,
        batch_size=32,
        verbose=0 # Set verbose to 0 to suppress training output
    )
}

print("Selected binary classification algorithms:")
for name in chosen_algorithms.keys():
    print(f"- {name}")
```

```python
for name, model in chosen_algorithms.items():
    print(f"Training {name}...")
    model.fit(X_train_scaled, y_train)
    print(f"{name} training complete.")
```

```python
# Dictionary to store predictions
model_predictions = {}

# Iterate through the trained models and make predictions on the test data
for name, model in chosen_algorithms.items():
    print(f"Making predictions with {name}...")
    y_pred = model.predict(X_test_scaled)
    model_predictions[name] = y_pred
    print(f"Predictions with {name} complete.")
    # Print first few predictions to verify
    print(f"First 10 predictions for {name}: {y_pred[:10]}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Dictionary to store evaluation results
evaluation_results = {}

# Iterate through the models and their predictions
for name, y_pred in model_predictions.items():
    print(f"Evaluating {name}...")

    # Calculate standard metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Calculate AUC-ROC if the model has predict_proba
    auc_roc = None
    if hasattr(chosen_algorithms[name], 'predict_proba'):
        y_pred_proba = chosen_algorithms[name].predict_proba(X_test_scaled)[:, 1]
        auc_roc = roc_auc_score(y_test, y_pred_proba)
    elif hasattr(chosen_algorithms[name], 'decision_function'): # For SVC
        y_pred_proba = chosen_algorithms[name].decision_function(X_test_scaled)
        auc_roc = roc_auc_score(y_test, y_pred_proba)


    # Store the metrics
    evaluation_results[name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-score": f1,
        "AUC-ROC": auc_roc
    }

    # Print the metrics
    print(f"  Accuracy: {accuracy:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall: {recall:.4f}")
    print(f"  F1-score: {f1:.4f}")
    if auc_roc is not None:
        print(f"  AUC-ROC: {auc_roc:.4f}")
    print("-" * 20)

# Display the evaluation results dictionary
print("\nSummary of Evaluation Results:")
display(evaluation_results)
```

```python
# Convert the evaluation_results dictionary to a DataFrame
evaluation_df = pd.DataFrame(evaluation_results).T

# Display the DataFrame
display(evaluation_df)
```

```python
print("Performance Summary of Binary Classification Models:")
display(evaluation_df)

# Identify the best performing model for each metric
best_accuracy_model = evaluation_df['Accuracy'].idxmax()
best_precision_model = evaluation_df['Precision'].idxmax()
best_recall_model = evaluation_df['Recall'].idxmax()
best_f1_model = evaluation_df['F1-score'].idxmax()
best_auc_roc_model = evaluation_df['AUC-ROC'].idxmax()

print("\nBest Performing Model for Each Metric:")
print(f"Accuracy: {best_accuracy_model} ({evaluation_df.loc[best_accuracy_model, 'Accuracy']:.4f})")
print(f"Precision: {best_precision_model} ({evaluation_df.loc[best_precision_model, 'Precision']:.4f})")
print(f"Recall: {best_recall_model} ({evaluation_df.loc[best_recall_model, 'Recall']:.4f})")
print(f"F1-score: {best_f1_model} ({evaluation_df.loc[best_f1_model, 'F1-score']:.4f})")
print(f"AUC-ROC: {best_auc_roc_model} ({evaluation_df.loc[best_auc_roc_model, 'AUC-ROC']:.4f})")

print("\nOverall Best Performing Models:")
# Consider models that perform well across multiple metrics, especially F1-score and AUC-ROC
top_models = evaluation_df.sort_values(by=['F1-score', 'AUC-ROC'], ascending=False)
display(top_models)

print("\nSummary of Model Performance:")
print(f"Logistic Regression showed the lowest performance across all metrics.")
print(f"Support Vector Machine had a perfect Recall but lower Precision compared to other top models.")
print(f"{best_random_forest = } and {best_gradient_boosting = } and {best_feedforward_neural_network = } models performed exceptionally well, achieving high scores in Accuracy, Precision, Recall, F1-score, and AUC-ROC.")
print(f"Based on F1-score and AUC-ROC, {best_f1_model} and {best_auc_roc_model} are the top performing algorithms.")
```
